```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from PIL import Image

# c nh v chuyn sang nh xm
image = Image.open('C:/Users/tungi/OneDrive/Documents/src_cv/img/01/face1.jpg').convert('L')

# Chuyn nh PIL sang mng NumPy
image_array = np.array(image)

# Tnh gradient theo hng x (o hm ring theo x)
gradient_x = ndimage.sobel(image_array, axis=1)

# Thit lp ngng
threshold = 200  # iu chnh gi tr ngng theo mong mun

# To nh nh phn t gradient
binary_image = (np.abs(gradient_x) > threshold).astype(np.uint8) * 255

# Hin th nh gc v ng bin
plt.subplot(121)
plt.imshow(image_array, cmap='gray')
plt.title("nh gc")

plt.subplot(122)
plt.imshow(binary_image, cmap='gray')
plt.title("ng bin")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from PIL import Image

# c nh v chuyn sang nh xm
image = Image.open('C:/Users/tungi/OneDrive/Documents/src_cv/img/01/face1.jpg').convert('L')

# Chuyn nh PIL sang mng NumPy
image_array = np.array(image)

# Tnh gradient theo hng x (o hm ring theo x) bng mt n Prewitt x
gradient_x = ndimage.convolve(image_array, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))

# Tnh gradient theo hng y (o hm ring theo y) bng mt n Prewitt y
gradient_y = ndimage.convolve(image_array, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))

# Tnh bin  gradient
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

# To li subplot vi 3 dng v 1 ct
fig, axs = plt.subplots(3, 1, figsize=(6, 10))

# Hin th nh gc, gradient theo hng x v bin  gradient
axs[0].imshow(image_array, cmap='gray')
axs[0].set_title("nh gc")

axs[1].imshow(gradient_x, cmap='gray')
axs[1].set_title("Gradient theo hng x")

axs[2].imshow(gradient_magnitude, cmap='gray')
axs[2].set_title("Bin  gradient")

plt.tight_layout()
plt.show()
```
